```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
# Load the dataset
data = pd.read_csv('/content/drive/MyDrive/human_activity_recognition_combined.csv')
```

```python
# Preprocessing
# Encoding categorical 'Activity' labels
label_encoder = LabelEncoder()
data['Activity'] = label_encoder.fit_transform(data['Activity'])
```

```python
# Separating features and labels
X = data.drop(['Subject', 'Activity'], axis=1)
y = data['Activity']
```

```python
# Scaling the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
# Function to plot confusion matrix
def plot_confusion_matrix(cm, classes, title):
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
    plt.title(title)
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
```

```python
# Artificial Neural Network Classifier
nn = MLPClassifier(random_state=42)
nn.fit(X_train, y_train)
nn_pred = nn.predict(X_test)
nn_accuracy = accuracy_score(y_test, nn_pred)
```

```python
# Decision Tree Classifier
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)
dt_pred = dt.predict(X_test)
dt_accuracy = accuracy_score(y_test, dt_pred)
```

```python
# Random Forest Classifier
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_pred)
```

```python
# Plotting Confusion Matrix for each model
plt.figure(figsize=(18, 5))
plt.subplot(1, 3, 1)
plot_confusion_matrix(confusion_matrix(y_test, nn_pred), label_encoder.classes_, 'ANN Confusion Matrix')
plt.subplot(1, 3, 2)
plot_confusion_matrix(confusion_matrix(y_test, dt_pred), label_encoder.classes_, 'Decision Tree Confusion Matrix')
plt.subplot(1, 3, 3)
plot_confusion_matrix(confusion_matrix(y_test, rf_pred), label_encoder.classes_, 'Random Forest Confusion Matrix')
plt.tight_layout()
plt.show()
```

```python
# Visualizing Feature Importances from Random Forest
feature_importances = pd.Series(rf.feature_importances_, index=X.columns)
top_features = feature_importances.nlargest(10)
plt.figure(figsize=(10, 6))
top_features.plot(kind='barh')
plt.title('Top 10 Feature Importances in Random Forest')
plt.show()
```

```python
# Evaluating and Printing Results
print("Neural Network Accuracy:", nn_accuracy)
print("Decision Tree Accuracy:", dt_accuracy)
print("Random Forest Accuracy:", rf_accuracy)
print("\nClassification Report for Random Forest:")
print(classification_report(y_test, rf_pred))
```
